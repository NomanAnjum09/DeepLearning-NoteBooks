```python
import numpy as np
import pandas as pd
import keras 
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=300,n_features=2,centers=2,cluster_std=1.0,random_state=101)
```

```python
X,y = data
```

```python
np.random.seed(seed=101)
z_noise = np.random.normal(size=len(X))
z_noise = pd.Series(z_noise)
```

```python
z_noise
```

```python
feat = pd.DataFrame(X)
```

```python
feat
```

```python
feat = pd.concat([feat,z_noise],axis=1)
```

```python
feat.columns = ['X1','X2','X3']
```

```python
feat.head()
```

```python
plt.scatter(feat['X1'],feat['X2'],c=y)
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
fig = plt.figure(figsize=(9,9))
```

```python
ax = fig.add_subplot(111,projection='3d')
ax.scatter(feat['X1'],feat['X2'],feat['X3'],c=y)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
encoder = Sequential()
encoder.add(Dense(2,activation='relu',input_shape=[3]))
```

```python
decoder = Sequential()
decoder.add(Dense(3,activation='relu',input_shape=[2]))
```

```python
autoEncoder = Sequential([encoder,decoder])
```

```python
autoEncoder.compile(loss='mse',optimizer=SGD(lr=0.5))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_Data = scaler.fit_transform(feat)
```

```python
autoEncoder.fit(scaled_Data,scaled_Data,epochs=10)
```

```python
encoded_2dim = encoder.predict(scaled_Data)
```

```python
encoded_2dim.shape
```

```python
plt.scatter(scaled_Data[:,0],scaled_Data[:,1],c=y)
plt.show()
```

```python
plt.scatter(encoded_2dim[:,0],encoded_2dim[:,1],c=y)
plt.show()
```

```python
result = decoder.predict(encoded_2dim)
```

```python
fig = plt.figure(figsize=(9,9))

ax = fig.add_subplot(111,projection='3d')
ax.scatter(result[:,0],result[:,1],result[:,2],c=y)
```
